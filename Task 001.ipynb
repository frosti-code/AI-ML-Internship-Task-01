```python
# --------------------------------------------
# Task 1: Exploring and Visualizing a Simple Dataset
# Objective:
# Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.
# Dataset: Iris Dataset
# --------------------------------------------

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------
# Step 1: Load the dataset
# --------------------------------------------
# Dataset can be loaded from local CSV or seaborn's built-in dataset
df = pd.read_csv("data/iris.csv")  # Change path if necessary

# --------------------------------------------
# Step 2: Basic inspection of the dataset
# --------------------------------------------
print("Shape of dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())

print("\nFirst 5 rows of Dataset:")
display(df.head())

print("\nDataset Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe())

# --------------------------------------------
# Step 3: Data Visualization
# --------------------------------------------

# Scatter Plot - Petal length vs Petal width
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="species",
    palette="Set2"
)
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.tight_layout()
plt.show()

# Histograms for all numeric columns
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, color="skyblue")
    plt.title(f"Histogram of {col}")
    plt.tight_layout()
    plt.show()

# Boxplots for all numeric columns
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col], color="lightcoral")
    plt.title(f"Boxplot of {col}")
    plt.tight_layout()
    plt.show()

# --------------------------------------------
# Step 4: Insights from Visualizations
# --------------------------------------------
# Example insights (modify based on your observation):
print("\nInsights:")
print("1. Setosa species has distinctly smaller petal lengths and widths compared to others.")
print("2. Versicolor and Virginica overlap slightly but Virginica tends to have larger petal dimensions.")
print("3. Histograms show that measurements are not perfectly normally distributed.")
print("4. Boxplots reveal potential outliers in petal width for Virginica species.")
```
